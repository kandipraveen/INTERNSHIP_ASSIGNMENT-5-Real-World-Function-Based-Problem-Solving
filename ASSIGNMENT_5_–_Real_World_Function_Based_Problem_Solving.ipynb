#Problem 1: Social Media – Post Engagement Analyzer


```python
def Post_Engagement_Analyzer(likes_list):
  total_likes = 0

  for i in likes_list:
    total_likes+=i

  if total_likes >= 1000:
    print("Total Likes :",total_likes)
    print("Post Status : Viral Post")
  else:
    print("Total Likes :",total_likes)
    print("Post Status : Normal Engagement")

likes_data=[200, 150, 300, 400]

Post_Engagement_Analyzer(likes_data)
```

#Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def check_rainfall_adequacy(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average_rainfall = total / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}")


# Example
check_rainfall_adequacy([70, 75, 68, 75], 70)
```

#Problem 4: Social Media – Duplicate Account Detection


```python
def detect_duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print(f"Duplicate Accounts Found: {result}")


# Example
detect_duplicate_accounts(["user1", "user2", "user1", "user3"])
```

#Problem 6: Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print(f"Premium Crops: {premium_crops}")


# Example
filter_premium_crops([1500, 2500, 1800, 3200])
```

#Problem 7: System Monitoring – Application Health Checker


```python
def check_system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")


# Example
check_system_health(7)
```

#Problem 8: Banking – Daily Transaction Limit Checker


```python
def check_transaction_limit(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")


# Example
check_transaction_limit(60000)
```

#Problem 9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:
            attended += 1

    attendance_percentage = (attended / total_classes) * 100

    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print(f"Attendance Percentage: {attendance_percentage}")
    print(f"Exam Eligibility: {eligibility}")


# Example
check_exam_eligibility([1, 1, 1, 0, 1])
```

#Problem 10: Smart Electricity Bill Analyzer


```python
def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example
total_bill, usage_status = calculate_electricity_bill(250)
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Status: {usage_status}")
```
